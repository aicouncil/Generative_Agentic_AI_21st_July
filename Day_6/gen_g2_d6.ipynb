```python
myfile = open('/content/IndiaUS.txt' , 'r')
mytext = myfile.read()
myfile.close()
```

```python
mytext
```

```python
#tokenization

from tensorflow.keras.preprocessing.text import Tokenizer
mytokenizer = Tokenizer()
mytokenizer.fit_on_texts([mytext])
word_index = mytokenizer.word_index
print(word_index)
```

```python
total_words = len(word_index) + 1
print("Vocabulary Size- " , total_words)
```

```python
lines = mytext.split("\n")
print(lines[:3])
```

```python
for i in range(1,5):
  print(i)
```

```python
my_input_sequences = []

for line in lines:
  #print(line)
  token_list = mytokenizer.texts_to_sequences([line])[0]
  for i in range(1,len(token_list)):
    #print(token_list[0:i+1])
    n_gram_sequences = token_list[0:i+1]
    my_input_sequences.append(n_gram_sequences)

print(my_input_sequences)
```

```python
sequence_lengths = []

for sequence in my_input_sequences:
  #print(len(sequence))
  sequence_lengths.append(len(sequence))

print(max(sequence_lengths))
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

input_sequences = pad_sequences(my_input_sequences , maxlen=83, padding='pre')
print(input_sequences)
```

```python
input_sequences[ : , :-1 ]         #array[row_slicing , column_slicing]
```

```python
import numpy as np

X = input_sequences[ : , :-1 ]
y = input_sequences[ : , -1: ]
```

```python
from keras.utils import to_categorical
Yt = to_categorical(y)
```

```python
Yt[0]
```

```python
print(X.shape)
print(y.shape)
```

```python
print(Yt.shape)
```

```python
from tensorflow.keras import models, layers

model = models.Sequential()
model.add(layers.Embedding(input_dim = total_words, output_dim = 16, input_length = 82))
model.add(layers.SimpleRNN(82))
model.add(layers.Dense(599 , activation = 'sigmoid'))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy' , metrics = ['accuracy'])
model.fit(X,Yt,epochs=100)
```
